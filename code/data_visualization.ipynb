Let's start with a plot of the variances for the multiple angles

```python
# import

import numpy as np
import os
import lecroy

from pathlib import Path
import matplotlib.pyplot as plt
```

```python
# read data
parent = Path(os.path.dirname(os.getcwd()))
date = "091027"
state = "cat1"

data_path = parent / "data" / "processed_data" / date / (state + '.npy')
df_cat1 = np.load(data_path)
df_cat1.shape
```

```python
os.getcwd()
```

```python
for angle in range(13):
    #all_traces = 
    
    plt.plot(df_cat1[angle, 0, :, :].var(axis=0), label=f'Angle {angle}')
plt.xlabel('Time Value')
plt.ylabel('Variance')
plt.title('Variance of Time Values for Each Angle')
plt.legend()
plt.show()
```
